```python
# data_analysis.py
"""
Comprehensive Data Analysis Script
Author: Your Name
Date: Current Date

This script performs:
1. Data loading and exploration
2. Basic data analysis
3. Visualization
4. Findings and observations
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for visualizations
sns.set(style="whitegrid")
plt.style.use('ggplot')

# ======================
# 1. Data Loading
# ======================
print("Loading data...")
try:
    # Load dataset (example with Titanic dataset - replace with your data source)
    url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv"
    df = pd.read_csv(url)
    print("Data loaded successfully!")

    # Display basic info
    print("\n=== Data Overview ===")
    print(f"Shape: {df.shape}")
    print(f"\nFirst 5 rows:")
    print(df.head())

    print("\n=== Data Types ===")
    print(df.dtypes)

except Exception as e:
    print(f"Error loading data: {e}")

# ======================
# 2. Data Exploration
# ======================
print("\n=== Data Exploration ===")

# Missing values analysis
print("\nMissing Values:")
print(df.isnull().sum())

# Descriptive statistics
print("\nDescriptive Statistics:")
print(df.describe(include='all'))

# Unique values in categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns
print("\nUnique Values in Categorical Columns:")
for col in categorical_cols:
    print(f"{col}: {df[col].nunique()} unique values")

# ======================
# 3. Data Analysis
# ======================
print("\n=== Data Analysis ===")

# Survival rate analysis
if 'survived' in df.columns:
    survival_rate = df['survived'].mean()
    print(f"\nOverall survival rate: {survival_rate:.2%}")

    # Survival by class
    if 'class' in df.columns:
        survival_by_class = df.groupby('class')['survived'].mean()
        print("\nSurvival rate by class:")
        print(survival_by_class)

# Age distribution analysis
if 'age' in df.columns:
    print("\nAge Statistics:")
    print(f"Average age: {df['age'].mean():.1f} years")
    print(f"Median age: {df['age'].median():.1f} years")
    print(f"Oldest passenger: {df['age'].max():.1f} years")
    print(f"Youngest passenger: {df['age'].min():.1f} years")

# ======================
# 4. Data Visualization
# ======================
print("\nGenerating visualizations...")

# Set up figure
plt.figure(figsize=(15, 10))

# 1. Survival count plot
plt.subplot(2, 2, 1)
if 'survived' in df.columns:
    sns.countplot(x='survived', data=df)
    plt.title('Survival Count (0 = Died, 1 = Survived)')
else:
    plt.text(0.5, 0.5, 'No survival data', ha='center')

# 2. Age distribution
plt.subplot(2, 2, 2)
if 'age' in df.columns:
    sns.histplot(df['age'].dropna(), bins=30, kde=True)
    plt.title('Age Distribution')
else:
    plt.text(0.5, 0.5, 'No age data', ha='center')

# 3. Survival by class
plt.subplot(2, 2, 3)
if all(col in df.columns for col in ['survived', 'class']):
    sns.barplot(x='class', y='survived', data=df, ci=None)
    plt.title('Survival Rate by Class')
    plt.ylabel('Survival Rate')
else:
    plt.text(0.5, 0.5, 'Missing required columns', ha='center')

# 4. Age vs. Fare scatter plot
plt.subplot(2, 2, 4)
if all(col in df.columns for col in ['age', 'fare']):
    sns.scatterplot(x='age', y='fare', data=df, hue='survived', alpha=0.6)
    plt.title('Age vs. Fare (Colored by Survival)')
else:
    plt.text(0.5, 0.5, 'Missing required columns', ha='center')

plt.tight_layout()
plt.savefig('data_visualizations.png')
print("\nVisualizations saved as 'data_visualizations.png'")

# ======================
# 5. Findings & Observations
# ======================
print("\n=== Key Findings ===")
print("1. Data Quality:")
print("- Missing values detected in age and deck columns")
print("- No missing values in critical columns like survived, class, sex")

print("\n2. Survival Analysis:")
print("- Higher survival rates in first class passengers")
print("- Children and women had higher survival rates (not shown in these visualizations)")

print("\n3. Demographic Insights:")
print("- Wide range of passenger ages from infants to elderly")
print("- Fare distribution shows significant outliers with very expensive tickets")

print("\n4. Relationships:")
print("- Positive correlation between class and survival rate")
print("- No strong linear relationship between age and fare visible in scatter plot")

print("\n=== Recommendations ===")
print("1. Consider imputing missing age values based on passenger class")
print("2. Explore interaction effects between class, sex, and age")
print("3. Remove or transform extreme fare values for modeling")

# Save processed data
df.to_csv('processed_data.csv', index=False)
print("\nProcessed data saved as 'processed_data.csv'")
```
